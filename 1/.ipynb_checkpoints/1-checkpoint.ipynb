```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show(img, is_gray=False):
    if is_gray:
        plt.imshow(img, cmap='gray')
    else:
        plt.imshow(img)
    
    plt.show()
    
def swap_r_b(img):
    temp = img.copy()
    
    red = img[:,:,2].copy()
    green = img[:,:,1].copy()
    blue = img[:,:,0].copy()
    
    temp[:,:,0] = red
    temp[:,:,2] = blue
    
    return temp

def mono_g(img):
    temp = img.copy()
    
    temp[:,:,0] = 0
    temp[:,:,2] = 0
    
    return temp

def mono_r(img):
    temp = img.copy()
    
    temp[:,:,0] = 0
    temp[:,:,1] = 0
    
    return temp

def gray(img):
    R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
    imgGray = 0.2989 * R + 0.5870 * G + 0.1140 * B
    
    return imgGray
    
def rep_pix(img1, img2):
    gray1, gray2 = gray(img1), gray(img2)
    
    centerx1, centery1 = [int(gray1.shape[0] / 2) - 50, int(gray1.shape[1] / 2) - 50]
    cent1 = gray1[centerx1:centerx1 + 100, centery1:centery1 + 100]
    
    centerx2, centery2 = [int(gray2.shape[0] / 2) - 50, int(gray2.shape[1] / 2) - 50]
#     cent2 = gray2[centerx2:centerx2 + 100, centery2:centery2 + 100]
    
    temp = gray2.copy()
    
    temp[centerx2:centerx2 + 100, centery2:centery2 + 100] += cent1
    
    return temp
    
imgs = ['fox.jpg', 'flower.jpg']

for img_path in imgs:
    img = mpimg.imread(img_path)
    show(img)
    
    show(swap_r_b(img))
    show(mono_g(img))
    show(mono_r(img))
    show(gray(img), True)

img1, img2 = mpimg.imread(imgs[0]), mpimg.imread(imgs[1])
show(rep_pix(img1, img2), True)

def get_min_max_mean_dev(img):
    img_g = mono_g(img)
    min_, max_, mean, dev = np.min(img_g), np.max(img_g), np.mean(img_g), np.std(img_g)
    
    return min_, max_, mean, dev
    
def sub_div_mul_add(img):
    min_, max_, mean, dev = get_min_max_mean_dev(img)
    
    temp = img.copy()
    temp = (((temp - mean) / dev) * 10) + mean
    
    return temp.astype(np.uint8)
    
def shift_left(img):
    img_g = mono_g(img)
    
    non = lambda s: s if s<0 else None
    mom = lambda s: max(0,s)

    ox, oy = -2, 0

    shift_img = np.zeros_like(img_g)
    shift_img[mom(oy):non(oy), mom(ox):non(ox)] = img_g[mom(-oy):non(-oy), mom(-ox):non(-ox)]
    
    return shift_img

def sub_shift_org(img):
    shift_img = shift_left(img)
    temp = img - shift_img
    
    return temp
    
def flip(img):
    return np.fliplr(img)
    
def change_intens(img):
    temp = img.copy()
    
    return 255 - temp
    
def noise(img):
    mean = 0
    var = 2000
    sigma = var ** 0.5
    gaussian = np.random.normal(mean, sigma, (img.shape[0],img.shape[1])) 
    noisy_image = np.zeros(img.shape, np.float32)

    noisy_image[:, :, 0] = img[:, :, 0] + gaussian
    noisy_image[:, :, 1] = img[:, :, 1] + gaussian
    noisy_image[:, :, 2] = img[:, :, 2] + gaussian

    return noisy_image
    
for img_path in imgs:
    img = mpimg.imread(img_path)
    min_, max_, mean, dev = get_min_max_mean_dev(img)
    print(min_, max_, mean, dev)
    
    show(sub_div_mul_add(img))
    show(shift_left(img))
    show(flip(img))
    show(change_intens(img))
    show(noise(img))
    
```
